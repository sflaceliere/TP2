```python
df = pd.read_csv("nodes-addresses.csv")
df

df_countries = pd.DataFrame()
df_countries["countries"] = df["countries"]

df_countries = df_countries.dropna(subset=["countries"])

df_countries["countries"].value_counts()
df_countries.sort_values("countries")
df_countries.to_csv('pays.csv', sep='\t', encoding='utf-8')

plt.figure(figsize=(30,16))
sns.countplot(x="countries", data=df)
plt.style.use('fivethirtyeight')
plt.xticks(rotation = 45, ha = 'right')
#sns.countplot(x="countries", data=df)
```
